Using the data generated above, calculate the following statistics without importing anything from the statistics module in the standard library (https://docs.python.org/3/library/statistics.html) and then comfirm your results match up to those that are obtained when using the statistics module (where possible):

````python
#Write a comment per statistical function
import random
random.seed(0)
salaries = [round(random.random()*1000000, -3) for _ in range(100)]
print(salaries)
````

````python
#Calculating the Mean Manually
mean = sum(salaries)/len(salaries) # The sum(salaries) adds all the values in the list altogether, whilst the len(salaries) finds the number of values in that list.
print("Mean: ", mean)

#Calculating the Mean using the imported Statistics Module
import statistics
mean2 = statistics.mean(salaries)
print("Mean using imported module: ", mean2)
````

````python
#Median
sortedSalaries = sorted(salaries) # We first sort the salaries so that the values are in ascending order
length = len(sortedSalaries) # Finds the length of the list
if length % 2 == 0:  # This means that the values are even, therefore no middle values, thus the need for the -1 to get the 2 values that comprise the middle
  median = (sortedSalaries[length//2 - 1] + sortedSalaries[length//2]) / 2
else:
  median = sortedSalaries[length // 2] # If the length of the salaries is odd, then that means there is a middle number. We just have to divide the sorted Salaries by 2 to get it.

print("Median: ", median)

#Median using the imported Statistics Module
import statistics
median2 = statistics.median(salaries)
print("Median using the module: ", median2)
````

````python
#Mode using Counter
from collections import Counter

counter = Counter(salaries) #This finds the number of times the value appeared in the list
print(counter)

max_count = max(counter.values()) #This finds the highest frequency present in the list
mode = [k for k, v in counter.items() if v == max_count]
#The k represents the values in the salary list, whilst the v represents the frequency of the said value
#This finds the value with the highest frequency in the list, or, the mode, as the v is == to max_count

print("Mode (Most Occuring Value):", mode)
print("Number of times occured: ", max_count)

#Mode using stat module
import statistics
mode1 = statistics.mode(salaries)
print("\nMode using module: ", mode1)
````

````python
# Sample Variance
salaryLength = len(salaries)  # To find the sample variance, we must first find the length of the list
mean = sum(salaries) / salaryLength  # Then calculate the mean of the salaries
squared_diff_sum = 0  # Then Initialize a variable to accumulate the squared differences

# We then Loop through each value in the salaries list and compute the sum of squared differences from the mean
for salary in salaries:
    squared_diff_sum += (salary - mean) ** 2

# Calculate the sample variance using the formula and divide by (n-1), or in this case, the length of the list minus 1
sample_variance = squared_diff_sum / (salaryLength - 1)

# Print the calculated sample variance
print("Sample Variance: ", sample_variance)

# Sample Variance using Statistics Module
import statistics
sample_variance1 = statistics.variance(salaries)
print("Sample Variance using statistics module: ", sample_variance1)
````

````python
#Sample Standard Deviation
standard_deviation = sample_variance ** 0.5 #We use the value of the sample_variance in the previous code to save up time and space
print("Standard Deviation: ", standard_deviation)

#Standard Deviation using the module
import statistics
standard_deviation1 = statistics.stdev(salaries)
print("Standard Deviation using the module: ", standard_deviation1)
````

###**Exercise 2**
Using the same data, calculate the following statistics using the functions in the statistics module where appropriate:
*  Range
*  Coefficient of variation
*  Interquartile range
*  Quartile coefficient of dispersion

````python
#Range
range = max(salaries) - min(salaries) #We get the range by subtracting the maximum value in the salaries list with the minimum value
print("Range: ", range)
````

````python
#Coefficient of Variation
import statistics

mean = statistics.mean(salaries) #To get the Coefficient of Variation, we must first find the mean
standard_deviation = statistics.stdev(salaries) #Then the standard deviation using the statistics module

coefficient_of_variation = (standard_deviation / mean) * 100 #Then use the formula to solve it

print("Coefficient of Variation: ", coefficient_of_variation)
````

````python
#Interquartile Range
quartile = statistics.quantiles(salaries) #This calculates the quartiles of the salaries list using the quantiles function

iqr = quartile[2] - quartile[0]
#After finding the quartiles, we subtracts the first and third quartiles to calculate the interquartile range (IQR)
#The values inside the bracket represent the quartile. 0 is first quartile whilst 2 is the third quartile
print("Quartiles:", quartile)
print("Interquartile Range: ", iqr) #822250.0 - 400500.0 = 421750
````

````python
#Quartile Coefficient of Dispersion
import statistics
qcd = (quartile[2] - quartile[0]) / (quartile[2] + quartile[0]) #By using the same principle as in the iqr, we just have to use the formula to find the qcd
rounded_qcd = round(qcd, 4) #This is just to round the output to 4 decimal places
print("Quartile Coefficient of Dispersion: ", qcd)
print("Rounded version: ", rounded_qcd)
````

````python
#9. Create a new dataframe "noDiabetes" that gathers data with no diabetes
noDiabetes = data[data['Outcome'] == 0]
noDiabetes
````

````python
#10. Create a new dataframe "Pedia" that gathers data with age 0 to 19
Pedia = data[(data['Age'] >= 0) & (data['Age'] <= 19)]
Pedia
````

````python
#11. Create a new dataframe "Adult" that gathers data with age greater than 19
Adult = data[data['Age'] > 19]
Adult
````

````python
#12. Use numpy to get the average age and glucose value
ave_age = np.mean(data['Age'])
ave_glc = np.mean(data['Glucose'])

print("Average Age:", ave_age)
print("Average Glucose Value:", ave_glc)
````

````python
#13. Use numpy to get the median age and glucose value.
med_age = np.median(data['Age'])
med_glc = np.median(data['Glucose'])

print("Median Age:", med_age)
print("Median Value of Glucose:", med_glc)
````

````python
#14. Use numpy to get the middle values of glucose and age.
sorted_age = np.argsort(data['Age']) #I first sorted the list so that I can accurately find the middle values
sorted_glc = np.argsort(data['Glucose'])

mid_age = len(sorted_age) // 2 #I then find the length of the list before doing doing floor division by 2 to find the middle values of the age and glucose
mid_glc = len(sorted_glc) // 2

middle_age_row = data.iloc[sorted_age[mid_age]] #I then used the .iloc function to find the values in the whole middle row of the middle age and glucose
middle_glc_row = data.iloc[sorted_glc[mid_glc]]
print("Middle row if Age is sorted:\n",middle_age_row, "\n\n")
print("Middle row if Glucose is sorted:\n",middle_glc_row)
#We can see that the result correlates with the output in task 13, where I got the median age and glucose value, if we compare these values
````
